### Defining Signals

In a general sense, a "signal" refers to a detectable and quantifiable event, condition, or information that carries some meaning. In various contexts, the term "signal" can have different specific meanings.

+ In the context of electrical engineering and electronics, a signal is typically a time-varying voltage or current that represents information. Signals can be analog or digital and are used to convey data, such as audio signals in a microphone, video signals in a television, or digital data in a computer.
+ In the context of telecommunications, signals are used to transmit information over long distances. This can include radio waves, electromagnetic signals, or digital signals over the internet. These signals can carry various types of data, such as voice, video, or text.

The concept of signals is fundamental in many fields, as they are a means of conveying information and can be analyzed, processed, and manipulated to extract meaningful insights or take appropriate actions. The interpretation of a signal often depends on the context in which it is used.

```python
import numpy as np 
import scipy as sp 
from matplotlib import pyplot as plt 

figsize = (12,2)
```

### Types of Signals

Signals can be categorized into various types based on different characteristics and properties. Here are some common types of signals:

+ **Continuous-time Signal:** A continuous signal is a type of signal that exists over an uninterrupted range of time or space.

```python
t = np.linspace(0, 5, 1000)

exp = 0.5 * np.exp(t)
signal = 10 * np.sin(2 * np.pi * t) + exp

plt.figure(figsize=figsize)
plt.plot(t, signal)
plt.show()
```

+ **Discrete-time Signal:**: A discrete signal is a type of signal that is defined only at specific, isolated points in time or space.

```python
t = np.linspace(0, 5, 50)
signal = np.sin(2 * np.pi * t)

plt.figure(figsize=figsize)
plt.stem(t, signal)
plt.show()
```

```python
# ploting diffents frequency sines 
t = np.linspace(0, 1.0, 2000)

f1 = 5 # hz 
f2 = 25 # hz

def sine_wave(t, frequency):
    return np.sin(2 * np.pi * frequency * t)

signal_5hz = sine_wave(t, f1)
signal_25hz = sine_wave(t, f2)
signal_combinated = signal_5hz + signal_25hz
```

```python
figure, axis = plt.subplots(3, sharex=True, figsize=figsize)

axis[0].plot(signal_5hz, color = 'green')
axis[0].set_title('5 Hz Signal')

axis[1].plot(signal_25hz, color = 'purple')
axis[1].set_title('25 Hz Signal')

axis[2].plot(signal_combinated, color = 'red')
axis[2].set_title('Result Signal')
```

### Statistics concepts applied to Signals

##### Mean or DC value (In electronics)
The mean ($\mu$) is the average value of a signal. It is found by adding all the samples together and then dividing the sum by the total number of samples (N):

$\mu = \frac{1}{N} \sum_{i=0}^{N-1} x_i$


```python
t = np.linspace(0, 100, 100)
signal = np.sin(2 * np.pi * 25 * t)
signal_mean = np.mean(signal)

plt.figure(figsize=figsize)
plt.plot(t, signal)
print(f'Signal mean: {signal_mean}')
```

```python
def calc_mean(signal):
    N = len(signal)
    sum = 0
    for i in range(N):
        sum += signal[i]
    return (1/N) * sum 

print(calc_mean(signal))
```

##### Variance
Variance is the average of the squared differences between each data point and the mean of the dataset. It measures the average spread of data points from the mean.

$\sigma ^2 = \frac{1}{N} \sum_{i=1}^N (x_i - \bar{x})$

where $\bar{x}$ represents the signal mean.

```python
signal_variance = np.var(signal)
print(f'Signal variance: {signal_variance}')
```

```python
def calc_variance(signal):
    N = len(signal)
    signal_mean = calc_mean(signal)
    sum = 0
    for i in range(N):
        sum += (signal[i] - signal_mean) ** 2 
    return sum / N

print(f'Signal variance: {calc_variance(signal)}')
```

##### Standard deviation
Standard deviation is the square root of the variance. It quantifies the average distance between data points and the mean. It is expressed in the same units as the data points.

$\sigma = \sqrt{\frac{1}{N} \sum_{i=1}^N (x_i - \bar{x})}$

where $\bar{x}$ represents the signal mean.

```python
signal_std = np.std(signal)
print(f'Signal standard deviation: {signal_std}')
```

```python
#def calc_std(signal):
#    return np.sqrt(calc_variance(signal))

def calc_std(signal):
    N = len(signal)
    mean = calc_mean(signal)
    sum = 0 
    for i in range(N):
        sum += (signal[i] - mean) ** 2
    return np.sqrt(sum / N)

print(f'Signal standard deviation: {calc_std(signal)}')
```
