### Linear systems
Basic definitions: 
+ **System:** A process that produces an output signal in response to an input signal
+ **Signal:** A signal describes how one parameter relates to another

### Sampling notation
+ **Continuous signals:** $x(t)$ | $y(t)$
+ **Discrete signals:** $x[n]$ | $y[n]$

### Information Encoding in an Analog waveform
+ **Time domain encoding:** Information is encoded in the sine waves of the signal
+ **Frequency domain encoding:** Information is encoded in the shape of the waveform

### Impulse decomposition
+ Impulse decomposition breaks an N samples signal into N components signals, each containing N samples.
+ Each of the component signals contains one point from the original signal. with the remainder of the values being zero.
