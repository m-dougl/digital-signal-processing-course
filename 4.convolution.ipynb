### **Convolution**
- Convolution is an essential operation used for various purposes, such as filtering, image processing, and feature extraction. It involves combining two signals to produce a third signal that represents the way one signal modifies the other. Convolution is particularly important when dealing with linear time-invariant (LTI) systems, as it describes the system's response to an input signal.

+ Convolution in discrete signals: $(x*h)[n]=\sum_{k=-\infty}^{\infty} x[k].h[n-k]$

+ Convolution in continuous signals: $(x*h)(t)=\int_{-\infty}^{\infty} x(\tau).h(t-\tau)d\tau$
