### Quantization
Quantization in signal processing is a process of mapping a continuous range of values to a discrete set of values. In other words, it involves representing a continuous signal, which can take on an infinite number of values, with a finite number of discrete values. This is done by dividing the continuous range into a finite number of intervals or levels and then assigning each input value to the closest level.

+ **S/H**: The sample and holding sub-module converts independent variables (x-axis values) from continuous to discrete
+ **Q**: The quantizer sub-module converts dependent variables (y-axis values) from continuous to discrete

### Sampling theorem (Nyquist-Shannon Sampling Theorem)
The sampling theorem states that a continuous signal can properly be sample only if it does NOT CONTAIN frequency compoments ($f$) ABOVE HALF OF THE sampling rate ($f_s$).

Example: If we are sampling at 50 hz (50 samples/second), the analog signal we are sampling must be made of frequencies from 25 hz (25cycles/second) and below.

| Sampling frequency| Maximum frequency in the signal|
|--|--|
| $f_s$ | $f_{max}$ |

Sampling theorem: $f_s \geq 2 f_{max}$

### Analog filters
+ **Passive Lowpass Filter:** This filter passes low frequencies and **block** high frequencies. It is constructed using only resistors and capacitors, it is also called RC passive lowpass filter.

- At high frequency **capacitor** becomes a **short circuit**
- At low frequency **capacitor** becomes a **open circuit**

where $f_c = \frac{1}{2 \pi RC}$
